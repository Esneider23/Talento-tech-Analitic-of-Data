```python
# import the library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create the dictionary with the data
data = {
    "year": [
        2003, 2004, 2005
    ],
    "income": [
        500, 700, 800
    ]
}

# Convert the dictionary into DataFrame

df = pd.DataFrame(data)

# Show the DataFrame

print(f"Dataframe \n {df}")


# Create a plot with the DataFrame

plt.plot(   
    df["year"],
    df["income"],
    'o:r'
)
# the x-axis is the year
plt.gca().set_xticks(df["year"])
plt.title("Income per year")
plt.xlabel("Year")
plt.ylabel("Income")
```
