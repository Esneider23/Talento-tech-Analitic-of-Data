```python
# Import the library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a DataFrame
data = {
    'A': [1,2,3,4,5],
    'B': [6,7,8,9,10],  
    'C': [11,12,13,14,15],
}

df = pd.DataFrame(data)

# print the DataFrame
print('Dataframe:\n', df)

# plot the DataFrame
plt.plot(df['A'], df['B'])
plt.xlabel('A')
plt.ylabel('B')
plt.title('A vs B')
plt.show()
```
